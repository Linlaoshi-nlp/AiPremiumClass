```python
from vllm import LLM, SamplingParams
from transformers import BitsAndBytesConfig
import torch


prompts = [
 "写一首关于秋天的诗。,",
    "用Python写一个简单的计算器程序。",
    "模仿鲁迅的文风写一段关于‘城市早高峰’的短文，体现对现实的观察和淡淡的讽刺。",
    "假设你是一名历史老师，学生问：‘如果荆轲刺秦成功，秦朝还会统一中国吗？’请给出条理清晰的分析。",
    "对比说明‘抑郁症’和‘普通情绪低落’的 3 个核心区别，以及何时需要寻求专业帮助。",
    "以‘午夜的旧书店突然亮起一盏从未见过的灯’为开头，写一个包含奇幻元素的短篇故事，结尾要有一个出人意料的反转。",
]
# 初始化
sampling_params = SamplingParams(temperature=0.8, top_p=0.95,     max_tokens=1500, )# 配置生成参数，设置最大生成长度（根据需求调整，如500 tokens）

llm = LLM(model="Qwen/Qwen-1_8B-Chat",    trust_remote_code=True, ) # 允许执行模型中的自定义代码  ##model='facebook/opt-125m'
# 进⾏推理
outputs = llm.generate(prompts, sampling_params)
# 打印输出
for output in outputs:
 prompt = output.prompt
 generated_text = output.outputs[0].text
 print(f"Prompt: {prompt!r}, Generated text: {generated_text!r}")
```

```python
from vllm import LLM, SamplingParams
prompts = [
    # 科技与未来
    "Artificial intelligence will revolutionize",
    "By 2030, quantum computing will",
    "The impact of blockchain on society is",
    
    # 文化与历史
    "The most significant invention of ancient China was",
    "Shakespeare's legacy can be described as",
    "Traditional Japanese tea ceremony represents",
    
    # 环境与自然
    "Climate change mitigation requires",
    "Coral reefs are essential because",
    "Sustainable agriculture practices include",
    
    # 哲学与思考
    "The meaning of life might be",
    "True happiness comes from",
    "Human consciousness is",
    
    # 创意与故事
    "In a distant galaxy, an alien explorer discovered",
    "The detective examined the mysterious object and realized",
    "As the clock struck midnight, the old mansion"
]

# 初始化模型和参数
sampling_params = SamplingParams(
    temperature=0.85,  # 稍高的温度增加创造性
    top_p=0.92,        # 更广的概率分布
    max_tokens=100,    # 生成长度
    frequency_penalty=0.5  # 减少重复
)

llm = LLM(model="facebook/opt-125m")

# 进行推理
outputs = llm.generate(prompts, sampling_params)

# 格式化输出
print("{:<60} {:<10}".format("Prompt", "Generated Text"))
print("="*100)
for output in outputs:
    prompt = output.prompt
    generated_text = output.outputs[0].text.strip()
    print("{:<60} {:<10}".format(f'"{prompt}"', f'"{generated_text[:80]}..."'))
```
